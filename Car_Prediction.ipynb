#Data Cleaning

Bringing some changes to the data set for better prediction

Converting object into integer for fuel type

---
1. CNG - 0
2. Diesel - 1
3. Petrol - 2


```
df.Fuel_Type.replace({'CNG':0,'Diesel':1,'Petrol':2},inplace = True)
```

Converting object to integer for seller type

---
1. Dealer - 1
2.Individual - 2

```
df.Seller_Type.replace({'Dealer':1,'Individual':2},inplace = True)
```

Converting object to integer in Transmiison type

---
1. Automatic - 1
2. Manual - 0

```
df.Transmission.replace({'Automatic':1,'Manual':0},inplace = True)
```

#Modelling

```
from sklearn import preprocessing,model_selection,svm
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor 
from sklearn.metrics import mean_squared_error,r2_score
```

```
df = df.drop('Car_Name' , axis = 1);
```

```
df.head()
```

```
y = pd.DataFrame(df['Selling_Price'])
X = (df.drop(['Selling_Price'] , axis = 1))
```

```
#Preprocessing the data.
X = preprocessing.scale(X)
```

```
#Splitting the data.
X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,test_size = 0.2)
```

## Linear Regression

```
clf = LinearRegression()

#Fitting the data.
clf.fit(X_train , y_train)

y_pred = clf.predict(X_test)

#Calculating Accuracy.
print("The R2 Score is {}".format(r2_score(y_test,y_pred)))
print("The Mean Squared Error is {}".format(mean_squared_error(y_test,y_pred)))
```

```
# create a regressor object 
regressor = DecisionTreeRegressor(random_state = 0)  
  
# fit the regressor with X and Y data 
regressor.fit(X, y) 

y_pred = regressor.predict(X_test)

#Calculating Accuracy.
print("The R2 Score is {}".format(r2_score(y_test,y_pred)))
print("The Mean Squared Error is {}".format(mean_squared_error(y_test,y_pred)))
```

The Linear Regression is giving an accuracy of approx 90%
